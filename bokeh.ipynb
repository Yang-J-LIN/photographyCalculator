```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cof(u, u2, f, F):
    return (u2 - u) / F * (f**2) / u2 / (u - f)
```

```python
distance_behind_object = np.linspace(1e-3, 50, 100)

focal_length_1 = 85 * 1e-3  # m
focal_length_2 = 200 * 1e-3 # m

object_distance_1 = 5     # m

# to keep the same size of the object in the image
object_distance_2 = object_distance_1 * focal_length_2 / focal_length_1

aperture_f_1 = 1.2
aperture_f_2 = 2.8


y = cof(object_distance_1, distance_behind_object + object_distance_1, focal_length_1, aperture_f_1)
y2 = cof(object_distance_2, distance_behind_object + object_distance_2, focal_length_2, aperture_f_2)
```

```python
plt.plot(distance_behind_object, y * 1e3, label='{} mm'.format(int(focal_length_1 * 1e3)))
plt.plot(distance_behind_object, y2 * 1e3, label='{} mm'.format(int(focal_length_2 * 1e3)))
plt.xlabel('Distance (m)')
plt.ylabel('Bokeh size (mm)')
plt.legend()
```
